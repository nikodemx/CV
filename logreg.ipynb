```python
from pyspark.sql import SparkSession
```

```python
from pyspark.ml.classification import LogisticRegression
```

```python
spark = SparkSession.builder.appName('logreg').getOrCreate()
```

```python
data = spark.read.format('libsvm').load(r"C:\Users\Asus\Desktop\studia big data\Python-and-Spark-for-Big-Data-master\Spark_for_Machine_Learning\Logistic_Regression\sample_libsvm_data.txt")
```

```python
logistic = LogisticRegression()
```

```python
fited = logistic.fit(data)
```

```python
fited_sum = fited.summary
```

```python
fited_sum.predictions.show()
```

```python
train_data, test_data = data.randomSplit([0.7, 0.3])
```

```python
train_data.show()
```

```python
fit_train = logistic.fit(train_data)
```

```python
correct = fit_train.evaluate(test_data)
```

```python
correct.predictions.show()
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
```

```python
eval = BinaryClassificationEvaluator()
```

```python
final = eval.evaluate(correct.predictions)
```
